Makine Öğrenmesi Nedir?

Bilgisayarların insanlara benzer şekilde öğrenmesini sağlamak maksadıyla çeşitli algoritma ve tekniklerin geliştirilmesi için çalışan bilimsel çalışma alanıdır.
***
Veri setindeki bağımlı değişken sayısal bir değişken ise problem regresyon problemidir.
*
Regresyon problemlerinde tahmin edilen değerler : y = b + wx
***
Bağımlı değişken bir kategorik değişken ise problem sınıflandırma problemi olacaktır.
***

Değişken Türleri Nelerdir?
* Sayısal Değişkenler
* Kategorik değişkenler (Nominal, Ordinal)
* Bağımlı Değişken (target, dependent, output, response)
* Bağımsız Değişken (feature, independent, input, column, predictor, explanatory)

***
Öğrenme Türleri Nelerdir?
* Denetimli Öğrenme (Supervised Learning)
* Denetimsiz Öğrenme(Unsupervised Learning)
* Pekiştirmeli Öğrenme(Reinforcement Learning)

***
Denetimli Öğrenme
* Eğer dataset içerisinde labellarımız mevcut ise bu denetimli örenmedir.
* Bir olay etkisi neticesinde sonuç değerleri dataset içerisinde yer alır.
* Target dataset içerisinde yer alır.

***

Denetimsiz Öğrenme
* Eğer dataset içerisinde labellar yok ise denetimsiz öğrenmedir.
* Bağımlı değişken dataset içerisinde yer almaz.
* Örn. müşteri segmentasyonu

***

Problem Türleri Nelerdir?
* Regresyon Problemi
** Eğer bağımlı değişken sayısal bir değişken ise problem regresyon problemidir.
* Sınıflandırma Problemi
** Eğer bağımlı değişken kategorik bir değişken ise problem sınıflandırma problemidir.

***

Model Başarı Değerlendirme Yöntemleri Nelerdir?

"Tahminler Ne Kadar Başarılırdır?"

* Regresyon Modellerinde: MSE - RMSE - MAE
* Sınıflandırma Modellerinde: Accuracy

***

Model Doğrulama Yöntemleri Nelerdir? (Model Validation)

* Holdout Yöntemi (Sınama Seti Yöntemi)
** Orjinal veri seti eğitim ve test seti olarak iki ye bölünür.
** Eğitim seti üstünde(train) model örenmesi gerçekleştirilir.
** Test Setinde ise model gerçek veriler ile test edilir.

* K-Katlı Çapraz Doğrulama (K Fold Cross Validation)
** Test seti hatasını minimize etmek için kullanılır.
** Orjinal veri seti 5 parçaya ayrılıp 4 set eğitim 1 set test olacak sekilde, parçaların kombinasyonu kadar train-test işlemi yapılabilir.
** İkinci yöntem olarak veri seti holdout gibi bölünüp eğitim seti k-katlı olarak parçalara bölünür. Eğitim seti cross validation yapılır sonrasında modelin hiç görmediği bir veri seti olan holdout test setinde de son test yapılır.

***

Yanlılık- Varyans Değiş Tokuşu (Bias - Variance Tradeoff)

* Aşırı öğrenme sorunları gözden geçirilir.
* OverFitting : Aşırı Öğrenme - Yüksek Varyans
** Overfitting : Modelin veriyi ezberlemesidir.

* Model Kurmak: Bağımlı ve bağımsız değişken arasındaki örüntüyü anlamak.

* Underfitting : Az öğrenme - Yüksek Yanlılık
** Underfitting : Modelin veriyi öğrenememesidir.

* Doğru veri: Düşük yanlılık , düşük varyans

* Aşırı öğrenmeye düşüldüğü nasıl anlaşılır?
** Eğitim seti ve test setinin model karmaşıklığı / tahmin hatasının birlikte değerlendirilmesi ile tespit edilir.

** Eğitim seti ile test seti arasındaki ilk ayrım kırılımı noktası optimum noktIDR.

***
Model karmaşıklığı nedir?
* Modelin hassaslaştırılması, daha dataylı tahminler yapabilmesi için özelliklerinin kuvvetlendirilmesidir.
* Model karmaşıklığının artması daha iyi model yaıldığını gösterir.

***
Aşırı öğrenme nasıl çözümlenir?
* Veri seti boyutu arttırılabilir.
* Feature selection yapılabilir.
* Eğer eğitim ve test setinin hatalarının birbirinden ayrılmaya başladığı nokta optimum nokta olarak belirlenip model orada durdurulur.
